**Search Operations**

```python
#In an unsorted array, the search operations can be performed by linear traversal(One-After_Another) from the first element to the last element.
def find_element(arr,key):
    for i in range(len(arr)):  #First traverse each element of the array
        if (arr[i]==key):      #Check if the value is equal to the key value or not
            return i           #If yes, return the index of the element
    return -1                  #If element not found, return -1
arr = [1,2,3,4,5]
key = 3
print(find_element(arr,key))
```

**Insert Operations**

***Insert at the End***

```python
#In an unsorted array, inserting an element is faster as compared to the sorted array because we don't need to care about any specific order.
def insert_end(arr, element):
    arr.append(element)        #Adds the element to the end of the arra
    return arr
arr = [1,2,3,4,5]
element = 6
print(insert_end(arr, element))
```

***Insert at any Position***

```python
#Insert at any position can be performed by shifting the elements to the right which are on the right side of the required position
def insert_elements(arr, n, x):
    arr.insert(n,x)            #Python has an inbuilt function called insert which takes the position and value for inserting the element at any particular index number
arr = [1,2,3,4,5]
n = 3
x = 7
if n>len(arr):                
    print("Enter valid index number")
else:
    insert_elements(arr, n, x) #Just call the function
    print(arr)                 #Just print the updated array
```

**Delete Operations**

```python
#To delete an element, we need to perform linear search and then if found, delete operations followed by shifting elements.
def delete_elements(arr, element):
    arr.remove(element)              #Python has a built-in function called 'remove' which removes the first occurences of the element from the postion and shifts the remaining elements.
arr = [1,2,3,4,5]
element = 4
delete_elements(arr, element)       #Just call the function 
print(arr)                          #Print the updated array
```
