```python
#Alternate elements
def findAlternates(arr):
    alternates = []
    for i in range(0,len(arr),2):
        alternates.append(arr[i])
    return alternates
arr = [1,2,3,4,5]
print(findAlternates(arr))
```

```python
#Recursive
def findAlternates(arr,index,res):
    if index < len(arr):
        res.append(arr[index])
        findAlternates(arr,index+2,res)
def recursionFindAlternates(arr):
    res = []
    findAlternates(arr,0,res)
    return res
arr = [1,2,3,4,5]
print(recursionFindAlternates(arr))
```

```python
#Finding leaders
def findLeaders(arr):
    res = []
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i]<arr[j]:
                break
            else:
                res.append(arr[i])
    return res
arr = [16,17,4,3,5,2]
print(findLeaders(arr))
```

```python
#Using Suffix Sum
def findLeaders(arr):
    res = []
    maxElement = arr[-1]
    res.append(maxElement)
    for i in range(len(arr)-2,-1,-1):
        if arr[i]>maxElement:
            maxElement = arr[i]
            res.append(maxElement)
    return res[::-1]
arr = [16,17,4,3,5,2]
print(findLeaders(arr))
```

```python
#Check if the array is sorted or not
#Iterative
def checkSort(arr):
    for i in range(len(arr)):
        if arr[i-1]>arr[i]:
            return False
    return True
arr = [16,17,4,3,5,2]
print(checkSort(arr))
```

```python
#Built-in Function
def built_in_Sorted(arr):
    return arr == sorted(arr)
arr = [16,17,4,3,5,2]
print(built_in_Sorted(arr))
```
